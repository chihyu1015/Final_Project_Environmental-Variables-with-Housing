```python
# import population, distance from the seaside file

ZIP_Centroid = pd.read_csv("ZIP_Centroid.csv")
ZIP_Centroid_selected = ZIP_Centroid[["ZIP_CODE","POPULATION","POP_SQMI","SQMI","NEAR_DIST"]]
ZIP_Centroid_renamed = (
    ZIP_Centroid_selected
    .rename(columns={"ZIP_CODE": "ZIP"})
    .rename(columns={"NEAR_DIST": "SEA_DIST"})
)

ZIP_Centroid_renamed.head()
```

```python
# import population, pollution, infectious diseases file

population = pd.read_csv("2018population.csv")
# CES 3.0 Score CalEnviroScreen 3.0: Composite Score (Higher indicates greater environmental burden)
# SB 535 Disadvantaged Community: Is it considered an underserved community under California SB 535 (Yes/No)
# Pesticides:  Pesticide Exposure Indicators
# Imp. Water Bodies:  Affected Water Body Indicators (Lake, River Pollution)
# Solid Waste:  Solid Waste-Related Indicators (e.g., Landfill Density)
# Pollution Burden: Comprehensive Percentage Indicator
# Asthma: Prevalence
# Low Birth Weight:   Proportion of Low Birth Weight Infants
# Cardiovascular Disease: Prevalence
# Education:  Proportion of the Population with Low Educational Attainment
# Linguistic Isolation: Proportion of Population with Limited or No English Proficiency in the Household
# Housing Burden: the proportion of housing costs, such as rent or mortgage payments, to household income
population_selected = population[[ "ZIP", "Total Population", "CES 3.0 Score", "SB 535 Disadvantaged Community", "PM2.5", "Drinking Water", 
                                  "Pesticides", "Traffic", "Groundwater Threats", "Imp. Water Bodies", "Solid Waste", "Pollution Burden", 
                                 "Asthma", "Cardiovascular Disease", "Education", "Linguistic Isolation", "Poverty", "Unemployment", "Housing Burden"]]
```

```python
# View missing values in 2018population.csv

missing_cols = population_selected.isna().mean() * 100
missing_cols = missing_cols[missing_cols >= 0].sort_values(ascending=False)

print("Columns containing missing values and their percentage of missing values (%):")
print(missing_cols)
```

```python
# Merge tables population_selected and ZIP_Centroid_filtered

invalid_zip = [32, 35, 39, 40, 48, 51, 57, 61]

population_selected = population_selected[~population_selected["ZIP"].isin(invalid_zip)]

population_selected = (
    population_selected
    .groupby("ZIP", as_index=False)
    .agg(lambda x: x.median() if pd.api.types.is_numeric_dtype(x) else x.iloc[0])
)


pop_combined = pd.merge(
    ZIP_Centroid_renamed,
    population_selected,
    on="ZIP",
    how="outer"
)

pop_combined["POPULATION"] = pop_combined["POPULATION"].fillna(pop_combined["Total Population"])

pop_combined = pop_combined.drop(columns=["Total Population"], errors="ignore")
```

```python
# Append new values to the price table

cols_exist = cols_exist[cols_exist["ZIP"].notna()]
pop_combined = pop_combined[pop_combined["ZIP"].notna()]

cols_exist["ZIP"] = cols_exist["ZIP"].astype(int)
pop_combined["ZIP"] = pop_combined["ZIP"].astype(int)

cols_merged = cols_exist.merge(
    pop_combined, 
    how="left",
    on="ZIP"
)

cols = ["ZIP"] + [c for c in cols_merged.columns if c != "ZIP"]
cols_merged = cols_merged[cols]
```

```python
cols_merged_nonull = cols_merged[cols_merged["POPULATION"].notna()].copy()
```

```python
# Divide the latitude and longitude columns by 1e6
cols_merged_nonull["latitude"] = cols_merged_nonull["latitude"] / 1e6
cols_merged_nonull["longitude"] = cols_merged_nonull["longitude"] / 1e6

cols_merged_nonull[["latitude", "longitude"]].head()
```
