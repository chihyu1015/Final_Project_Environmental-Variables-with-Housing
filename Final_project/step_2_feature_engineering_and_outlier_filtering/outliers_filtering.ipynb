```python
import pandas as pd
import numpy as np
import os

df = pd.read_csv("selected_columns.csv")
cols = ['taxvaluedollarcnt', 'finishedsquarefeet12', 'lotsizesquarefeet']

for c in cols:
    df[c + '_log'] = np.log1p(df[c])

mask_sigma = pd.Series(True, index=df.index)

for c in cols:
    log_col = c + '_log'
    mean_val = df[log_col].mean()
    std_val = df[log_col].std()
    lower = mean_val - 4 * std_val
    upper = mean_val + 4 * std_val
    mask_sigma &= (df[log_col] >= lower) & (df[log_col] <= upper)

df_clean = df[mask_sigma].copy()
df_clean.drop(columns=[c + '_log' for c in cols], inplace=True)

save_dir = r"C:\\Users\\caiyu\\Documents\\DSCI\\DSCI550\\final project\\final_document"
os.makedirs(save_dir, exist_ok=True)
save_path = os.path.join(save_dir, "dsci_filter_log_4sigma.csv")
df_clean.to_csv(save_path, index=False)

print("Rows before: ", len(df))
print("Rows after: ", len(df_clean))
print("Saved to: ", save_path)
```
